```python
from twnews.dataset.url_resolver import UrlResolver
url_resolver = UrlResolver()
k = url_resolver.resolve_url_map.keys()

urls = []
for tweet in tweets.get_dataset_texts():
    tweet.resolve_urls(url_resolver)
    urls.extend(tweet.urls)
```

```python
print len(tweets.dataset_texts_dict)
print len(urls)
print len(set(urls))
from urlparse import urlparse
def get_domain(url):
    return urlparse(url)[1]
url_to_domain = {url:get_domain(url) for url in urls}
```
